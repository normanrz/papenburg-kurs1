```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
mean1 = [0, 0]
cov1 = [[1, 0], [0, 1]]
mean2 = [4, 4]
cov2 = [[1, 1], [0, 1]]
x1 = np.random.multivariate_normal(mean1, cov1, 5000)
x2 = np.random.multivariate_normal(mean2, cov2, 5000)
data = np.concatenate((x1, x2), axis=0)
np.random.shuffle(data)

plt.plot(data[:,0], data[:,1],'xb')
plt.show()
```

```python
kMeans = KMeans(n_clusters=2)
kMeans.fit(data)
print kMeans.cluster_centers_
for i in range(0, 2):
    print "%d: %d" % (i, np.sum(kMeans.labels_ == i))
```

```python
for i in range(0, 2):
    plt.plot(data[kMeans.labels_ == i, 0], data[kMeans.labels_ == i, 1], 'x')
plt.plot(kMeans.cluster_centers_[:,0], kMeans.cluster_centers_[:,1],'xr')
plt.show()
```
