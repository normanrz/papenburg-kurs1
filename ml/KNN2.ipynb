```python
%matplotlib inline

import numpy as np
from sklearn import datasets
from time import sleep
import matplotlib.pyplot as plt
```

```python
# Hier sind ein paar Beispielzahlendaten
digits = datasets.load_digits()
plt.figure(1, figsize=(10, 10))
for x in range(20):
  plt.subplot(4, 5, x + 1)
  plt.imshow(digits.images[x], cmap=plt.cm.gray_r, interpolation='nearest')
```

```python
def knn_predict(trainX, trainY, testX):
    return trainY[np.argmin(np.sum((trainX - testX) ** 2, axis=1))]
```

```python
# Wir holen die ersten 1000 Bilder als Trainingsdaten
trainingDigits = digits.images[0:1000]
# Außerdem machen wir die Bilder flach, 1000 Zeilen Daten also.
trainingDigitsFlat = trainingDigits.reshape(1000, -1)
trainingLabels = digits.target[0:1000]


# Nun klassifizieren wir 10 Zahlen
plt.figure(2, figsize=(16, 2))   
for i in range(1200, 1210):
    
    digit = digits.images[i]
    digitFlat = digit.reshape(1, -1)
    digitLabel = digits.target[i]
    
    plt.subplot(1, 10, i - 1200 + 1)
    plt.imshow(digit, cmap=plt.cm.gray_r, interpolation="nearest")
    
    print knn_predict(trainingDigitsFlat, trainingLabels, digitFlat), digitLabel
```
